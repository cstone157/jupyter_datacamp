```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

volunteer = pd.read_csv("./data/volunteer_opportunities.csv")
```

```python
# Drop the Latitude and Longitude columns from volunteer
volunteer_cols = volunteer.drop(['Latitude', 'Longitude'], axis=1)

# Drop rows with missing category_desc values from volunteer_cols
volunteer_subset = volunteer_cols.dropna(subset=["category_desc"])

# Print out the shape of the subset
print(volunteer_subset.shape)
```

```python
# Print the head of the hits column
print(volunteer["hits"].head())

# Convert the hits column to type int
volunteer["hits"] = volunteer["hits"].astype("int")

# Look at the dtypes of the dataset
print(volunteer.dtypes)
```

```python
#volunteer = volunteer.dropna()
#volunteer.columns

volunteer = volunteer[['opportunity_id', 'content_id', 'vol_requests', 'event_time', 'title', 'hits', 'summary', 'category_id', 'category_desc', 'org_title', 'org_content_id',
       'addresses_count', 'locality', 'region', 'postalcode', 'display_url', 'recurrence_type', 'hours', 'created_date', 'last_modified_date', 'start_date_date', 'end_date_date',
       'status']]

volunteer = volunteer.dropna()
volunteer.head()
```

```python
# Create a DataFrame with all columns except category_desc
X = volunteer.drop("category_desc", axis=1)

# Create a category_desc labels dataset
y = volunteer[["category_desc"]]

# Use stratified sampling to split up the dataset according to the y dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)

# Print the category_desc counts from y_train
print(y_train["category_desc"].value_counts())
```
